```python
class ProbDist(dict):
    """A Probability Distribution; an {outcome: probability} mapping."""
    def __init__(self, mapping=(), **kwargs):
        self.update(mapping, **kwargs)
        # Make probabilities sum to 1.0; assert no negative probabilities
        total = sum(self.values())
        for outcome in self:
            self[outcome] = self[outcome] / total
            assert self[outcome] >= 0
```

```python
def p(event, space): 
    """The probability of an event, given a sample space of equiprobable outcomes. 
    event: a collection of outcomes, or a predicate that is true of outcomes in the event. 
    space: a set of outcomes or a probability distribution of {outcome: frequency} pairs."""
    # branch on the type of the first argument
    if is_predicate(event):
        # transform the mapping (untangible) 'event' into the collection (tangible) 'event'
        event = such_that(event, space)
        
    if isinstance(space, ProbDist):
        # if space is a dictionary of distinct probabilities, where each item does not count as the same amount
        return sum(space[o] for o in space if o in event)
    else:
        # space is not a dictionary but a collection, let's fall back to our original division
        return Fraction(len(event & space), len(space))

is_predicate = callable

def such_that(predicate, space): 
    """The outcomes in the sample pace for which the predicate is true.
    If space is a set, return a subset {outcome,...} with outcomes where predicate(element) is true;
    if space is a ProbDist, return a ProbDist {outcome: frequency,...} with outcomes where predicate(element) is true."""
    if isinstance(space, ProbDist):
        return ProbDist({o:space[o] for o in space if predicate(o)})
    else:
        return {o for o in space if predicate(o)}
```

#### If Red Bull wins at least one of these three races, what is the probability Red Bull wins all three?

```python
# Given probabilities of Red Bull winning each race
def redbull_atleast_one_win():
# Probability of Red Bull winning none of the three races
    prob_no_win = (1 - (PlayerProb1['MV'] + PlayerProb1['SP'])) * (1 - (PlayerProb1['MV'] + PlayerProb1['SP'])) * (1 - (PlayerProb1['MV'] + PlayerProb1['SP']))

# Probability of Red Bull winning at least one of the three races
    prob_at_least_one_win = 1 - prob_no_win

    return prob_at_least_one_win
```

```python
def redbull_win_all_races():
# Probability of Red Bull winning all three races
    prob_all_wins = (PlayerProb1['MV'] + PlayerProb1['SP']) * (PlayerProb1['MV'] + PlayerProb1['SP']) * (PlayerProb1['MV'] + PlayerProb1['SP'])

    return prob_all_wins
```

```python
'''P(a) = Red bull wins all races
P(b)= Red bull wins atleast one race
Using Bayes Theorem'''

prob_b = redbull_atleast_one_win()
prob_a = redbull_win_all_races()
prob_b_given_a = 1 #Redbull has won atleast 1 of the race
prob_a_given_b = (prob_b_given_a * prob_a)/prob_b
prob_a_given_b
```

```python
ratings_2023 = {
    "R_Verstappen": 575,
    "R_Perez": 285,
    "M_Hamilton": 234,
    "A_Alonso": 206,
    "F_Leclerc": 206,
    "MC_Norris": 205,
    "F_Sainz": 200,
    "M_Russell": 175,
    "MC_Piastri": 97,
    "A_Stroll": 74,
    "AR_Gasly": 62,
    "AR_Ocon": 58,
    "W_Albon": 27,
    "AH_Tsunoda": 17,
    "AF_Bottas": 10,
    "H_Hulkenberg": 9,
    "AH_Ricciardo": 6,
    "ALF_Guanyu": 6,
    "H_Magnussen": 3,
    "AH_Lawson": 2,
    "W_Sargeant": 1,
    "AH_DeVries": 0
}
```

```python
# Checking Ferrari's performance
def ferrari_performance(ratings):
    ferrari_drivers = ["F_Leclerc", "F_Sainz"]
    
    # Calculate the existing total rating for Ferrari 
    rating_without_hamilton = sum(ratings[driver] for driver in ferrari_drivers)
    print(rating_without_hamilton)
    
    # Adding Hamilton's rating to Ferrari's existing drivers rating
    rating_with_hamilton = rating_without_hamilton + ratings["M_Hamilton"]
    print(rating_with_hamilton)
    
    # Probability of Ferrari winning without Hamilton 
    ferrari_win_without_hamilton = rating_without_hamilton / sum(ratings.values())
    
    # Probability of Ferrari winning with Hamilton
    ferrari_win_with_hamilton = rating_with_hamilton / sum(ratings.values())
    
    return ferrari_win_without_hamilton, ferrari_win_with_hamilton

```

```python
# Calculating Ferrari's rating with and without Hamilton
prob_without_hamilton, prob_with_hamilton = ferrari_performance(ratings_2023)
print("Probability of Ferrari winning without Hamilton:", prob_without_hamilton)
print("Probability of Ferrari winning with Hamilton:", prob_with_hamilton)

'''Given the driver ratings of Hamilton, chances of winning definitely increase because his ratings is higher 
than both Leclerc and Sainz. It is the driver and skills that are more important for winning a race.'''
```
